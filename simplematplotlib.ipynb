```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[2,3,5,7,11]
```

```python
plt.plot(x,y)
plt.title('prime number plot ')
plt.xlabel('label')
plt.ylabel('prime number ')
plt.show()
```

```python
y1=[2,3,5,7,11]
z1=[1,4,6,8,12]
```

```python
plt.plot(x,y1,label='primenumber')
plt.plot(x,z1,label='random number')
plt.title('graph plot between different numbers')
plt.legend()
plt.show()
```

```python
### customizing these datasets
```

```python
plt.plot(x, y1, color='r', linestyle='--', label='primenumber')
plt.plot(x, z1, color='g', linestyle=':', label='randomnumber')
plt.legend()
plt.show()
```

```python
categories=['A','B','C','D','E','F']
values=[2,5,6,8,3,10]
plt.bar(categories,values)


plt.show()
```

```python
plt.plot(x, y)
plt.grid(False)
plt.annotate('prime at (3,5)',
             xy=(3, 5),
             xytext=(4, 6),
             arrowprops=dict(facecolor='black'))
plt.show()
```
